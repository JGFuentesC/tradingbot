```python
import aiohttp
from google.oauth2.service_account import Credentials
from google.auth.transport.requests import Request
import google.oauth2.id_token
import pandas as pd

import os
import asyncio
```

```python
url = 'https://us-central1-m5g18-397001.cloudfunctions.net/trading-dev'
```

```python
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = '/home/jose/Descargas/m5g18.json'
google_auth_request = Request()
id_token = google.oauth2.id_token.fetch_id_token(
    google_auth_request, url)
```

```python
len(id_token)
```

```python
async def call_gcf(url, data=None):
    print("Ejecutando...",data)
    headers = {
        'Authorization': 'Bearer ' + id_token
    }

    async with aiohttp.ClientSession() as session:
        async with session.post(url, headers=headers, json=data) as response:
            return await response.json()


```

```python
snp = pd.read_pickle('snp.pkl')
```

```python
stocks = snp['stock'].tolist()
```

```python
asyncio.gather(*map(lambda stock:call_gcf(url, {"stock":stock}),stocks))
```
